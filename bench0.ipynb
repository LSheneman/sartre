# Acceptance testing

```python
import numpy
shortdata = numpy.loadtxt('short/log.out')
```

```python
plot(shortdata[:,0], shortdata[:,1])
xlabel('time (s)')
ylabel('CPU load (% total CPUs)')
title("CPU usage")
axis(ymax=130)

annotate("QC start", (0, 50))
annotate("diginorm start", (521, 70))
annotate("assembly start", (1418, 100))
annotate("TF start", (1750, 120))
annotate("expr", (1753, 110))
annotate("DE", (2130, 110))
```

```python
import numpy
plot(shortdata[:,0], shortdata[:,2])
xlabel('time (s)')
ylabel('RAM (kb)')
title("Memory usage")

annotate("QC start", (0, 1.0e7))
annotate("diginorm start", (525, 1.1e7))
annotate("assembly start", (744, 1e7))
annotate("TF start", (1061, 0.9e7))
annotate("expr", (1507, 1e7))
annotate("DE", (2301, 1.1e7))
```
