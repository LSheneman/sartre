```python
import numpy
cpu = numpy.loadtxt('out')
```

```python
plot(cpu[:,0], cpu[:,1])
xlabel('time (s)')
ylabel('CPU load (% total CPUs)')
title("CPU usage")
axis(ymax=130)

annotate("QC start", (0, 50))
annotate("diginorm start", (525, 70))
annotate("assembly start", (744, 100))
annotate("TF start", (1061, 120))
annotate("expr", (1507, 110))
annotate("DE", (2301, 110))
```
