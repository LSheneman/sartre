```python
import numpy
logdata = numpy.loadtxt('log.out')
```

```python
plot(logdata[:,0], logdata[:,1])
xlabel('time (s)')
ylabel('CPU load (% total CPUs)')
title("CPU usage")
axis(ymax=130)

annotate("QC start", (0, 50))
annotate("diginorm start", (525, 70))
annotate("assembly start", (744, 100))
annotate("TF start", (1061, 120))
annotate("expr", (1507, 110))
annotate("DE", (2301, 110))
```

```python
import numpy
plot(logdata[:,0], logdata[:,2])
xlabel('time (s)')
ylabel('RAM (kb)')
title("Memory usage")

annotate("QC start", (0, 1.0e7))
annotate("diginorm start", (525, 1.1e7))
annotate("assembly start", (744, 1e7))
annotate("TF start", (1061, 0.9e7))
annotate("expr", (1507, 1e7))
annotate("DE", (2301, 1.1e7))
```
