```python
def triptych(data):
    # Two subplots, the axes array is 1-d
    f, axarr = plt.subplots(3, sharex=True)
    plot1, plot2, plot3 = axarr
    plot1.plot(data[:,0], data[:,1], 'b-')
    plot1.set_ylabel('CPU load (%)')
    
    #axarr[0].set_title('Sharing X axis')
    plot2.plot(data[:,0], data[:,2], 'r-')
    plot2.set_ylabel('RAM (GB)')
    plot3.plot(data[:,0], data[:,3], 'g-')
    plot3.set_ylabel('Disk (TPS)')
                     
    plot3.set_xlabel('Time (seconds)')
    
    f.set_size_inches(10, 8)
   
    return f, axarr
    
import numpy
shortdata = numpy.loadtxt('short-m1.xlarge/log.out')

f, axarr = triptych(shortdata)
plot1 = axarr[0]

plot1.set_ylim((0, 130))
plot1.annotate("QC start", (0, 50))
plot1.annotate("diginorm start", (521, 70))
plot1.annotate("assembly start", (1418, 100))
plot1.annotate("TF start", (1750, 120))
plot1.annotate("expr", (1753, 110))
plot1.annotate("DE", (2130, 110))
savefig('/tmp/fig1.pdf')
```
